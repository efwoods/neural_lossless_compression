# Data Exploration and Visualization

Visualizing samples of the data to better understand the dataset

```python
import matplotlib.pyplot as plt
from scipy.io import wavfile
import os
```

```python
data_folder = "./data/"
```

```python
# grab name of one file in the folder
data_dir = os.listdir("./data/")
example_filepath = os.path.join(data_folder, data_dir[0])
```

```python
sample_rate, data = wavfile.read(example_filepath)
sample_rate, data.shape
# 98699 / 19531 = 5.0534534842045975 seconds of recordings
```

```python
plt.figure(figsize=(10, 4))
plt.plot(data)
plt.title('Waveform of the Audio File')
plt.xlabel('Sample Index')
plt.ylabel('Amplitude')
plt.show()
```

```python
plt.figure(figsize=(10, 4))
plt.plot(data[:400])
plt.title('Waveform of the Audio File')
plt.xlabel('Sample Index')
plt.ylabel('Amplitude')
plt.show()
```
